```python
import os
import warnings
import numpy as np
import pandas as pd
import polars as pl
from tqdm import tqdm
import pickle
import json
from sklearn.preprocessing import LabelEncoder
from datetime import datetime
warnings.filterwarnings('ignore')

# Install AutoGluon if not already installed (should be handled by a separate cell usually)
# %pip install -U autogluon

from autogluon.tabular import TabularPredictor

def set_seed(seed=42):
    """Set random seed for reproducibility"""
    np.random.seed(seed)
    import random
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)

set_seed()
DATA_DIR = "/content/" # Use /content/ as the base data directory
MODELS_DIR = "/kaggle/working/autogluon-models/" # Save trained models to working directory
OUTPUT_DIR = "/kaggle/working/"

def process_data_for_autogluon(df: pl.DataFrame) -> pl.DataFrame:
    """Process data and add engineered features for AutoGluon"""
    # Add temporal features (using the existing function)
    df = add_temporal_features(df)

    # Add inter-player features (using the existing function)
    df = calculate_inter_player_features(df)

    # Convert player_height to numerical (assuming format 'feet-inches')
    if 'player_height' in df.columns:
        df = df.with_columns(
            (pl.col('player_height').str.split('-').list.get(0).cast(pl.Float64) * 12 +
             pl.col('player_height').str.split('-').list.get(1).cast(pl.Float64)
            ).alias('player_height_total_inches') # Calculate total inches and alias
        )


    # Convert player_birth_date to age (assuming format 'YYYY-MM-DD')
    if 'player_birth_date' in df.columns:
         df = df.with_columns(
             (pl.datetime(2024, 1, 1) - pl.col("player_birth_date").str.strptime(pl.Date, "%Y-%m-%d")).dt.total_days().alias("player_age_days")
         )


    # Drop potentially problematic columns for AutoGluon or those that won't be used
    cols_to_drop = ['player_name', 'player_birth_date', 'player_height'] # player_name is high cardinality, birth_date is handled, original player_height is now numerical
    df = df.drop([col for col in cols_to_drop if col in df.columns])

    return df

def add_temporal_features(df: pl.DataFrame) -> pl.DataFrame:
    """Add engineered temporal features that change across frames"""
    # Check if required columns exist
    if 'dir' not in df.columns or 's' not in df.columns:
        print("Warning: 'dir' or 's' columns missing for velocity calculation")
        return df

    # Velocity components (following data convention)
    df = df.with_columns([
        (pl.col("s") * (pl.col("dir") * np.pi / 180).sin()).alias("speed_x"),
        (pl.col("s") * (pl.col("dir") * np.pi / 180).cos()).alias("speed_y")
    ])

    # Momentum features using player weight converted to kg (lbs / 2.20462)
    if 'player_weight' in df.columns:
        df = df.with_columns([
            (pl.col("speed_x") * (pl.col("player_weight").fill_null(200) / 2.20462)).alias("momentum_x"),
            (pl.col("speed_y") * (pl.col("player_weight").fill_null(200) / 2.20462)).alias("momentum_y")
        ])
    else:
        # Fallback using average NFL player weight in kg (~90.7 kg = 200 lbs)
        df = df.with_columns([
            (pl.col("speed_x") * 90.7).alias("momentum_x"),
            (pl.col("speed_y") * 90.7).alias("momentum_y")
        ])

    # Add acceleration components
    if 'a' in df.columns and 'dir' in df.columns:
         df = df.with_columns([
            (pl.col("a") * (pl.col("dir") * np.pi / 180).sin()).alias("accel_x"),
            (pl.col("a") * (pl.col("dir") * np.pi / 180).cos()).alias("accel_y")
        ])

    # Add speed-acceleration interaction
    if 's' in df.columns and 'a' in df.columns:
        df = df.with_columns((pl.col("s") * pl.col("a")).alias("speed_accel_interaction"))

    # Add directional changes
    if 'dir' in df.columns:
        df = df.with_columns(
            pl.col("dir").diff().over(["game_id", "play_id", "nfl_id"]).fill_null(0).alias("dir_change")
        )

    return df

def calculate_inter_player_features(df: pl.DataFrame) -> pl.DataFrame:
    """Calculate features related to interactions between players"""
    # Ensure required columns exist
    required_cols = ["game_id", "play_id", "nfl_id", "frame_id", "x", "y", "player_side"]
    if not all(col in df.columns for col in required_cols):
        print("Warning: Missing required columns for inter-player features")
        return df

    # Separate offense and defense
    df_offense = df.filter(pl.col("player_side") == "Offense").select(["game_id", "play_id", "frame_id", "nfl_id", "x", "y"])
    df_defense = df.filter(pl.col("player_side") == "Defense").select(["game_id", "play_id", "frame_id", "nfl_id", "x", "y"])

    # Rename columns for easier merging
    df_offense = df_offense.rename({"nfl_id": "off_nfl_id", "x": "off_x", "y": "off_y"})
    df_defense = df_defense.rename({"nfl_id": "def_nfl_id", "x": "def_x", "y": "def_y"})

    # Cross join offense and defense within each frame, play, and game
    # This creates all pairs of offensive and defensive players for each moment
    inter_player = df_offense.join(
        df_defense,
        on=["game_id", "play_id", "frame_id"],
        how="inner"
    )

    # Calculate distance between offensive and defensive players
    inter_player = inter_player.with_columns(
        ((pl.col("off_x") - pl.col("def_x")).pow(2) + (pl.col("off_y") - pl.col("def_y")).pow(2)).sqrt().alias("distance_off_def")
    )

    # For each player, find the minimum distance to any player on the opposing side
    min_distance_agg = inter_player.group_by(["game_id", "play_id", "frame_id", "off_nfl_id"]).agg(
        pl.col("distance_off_def").min().alias("min_distance_to_defense")
    ).rename({"off_nfl_id": "nfl_id"}) # Rename back to nfl_id

    min_distance_agg_def = inter_player.group_by(["game_id", "play_id", "frame_id", "def_nfl_id"]).agg(
        pl.col("distance_off_def").min().alias("min_distance_to_offense")
    ).rename({"def_nfl_id": "nfl_id"}) # Rename back to nfl_id

    # Merge the minimum distances back into the original dataframe
    df = df.join(min_distance_agg, on=["game_id", "play_id", "frame_id", "nfl_id"], how="left")
    df = df.join(min_distance_agg_def, on=["game_id", "play_id", "frame_id", "nfl_id"], how="left")

    # Fill nulls created by players having no opponent on the field at a given frame
    df = df.with_columns([
        pl.col("min_distance_to_defense").fill_null(999.0), # Use a large value if no defense is present
        pl.col("min_distance_to_offense").fill_null(999.0) # Use a large value if no offense is present
    ])


    # Calculate relative position and angle to the ball landing location for each player
    if all(col in df.columns for col in ['ball_land_x', 'ball_land_y', 'x', 'y']):
        df = df.with_columns([
            ((pl.col("ball_land_x") - pl.col("x")).pow(2) +
             (pl.col("ball_land_y") - pl.col("y")).pow(2)).sqrt().alias("distance_to_ball_landing_frame"),

            pl.arctan2(pl.col("ball_land_y") - pl.col("y"),
                      pl.col("ball_land_x") - pl.col("x")).alias("angle_to_ball_landing_frame")
        ])

    return df

def prepare_training_data(df_input_train: pl.DataFrame, df_output_train: pl.DataFrame) -> pd.DataFrame:
    """Prepare training data by merging input and output and adding features"""
    print("Preparing training data...")

    # Apply feature engineering to training input data
    df_input_train_processed = process_data_for_autogluon(df_input_train)

    # Before merging, rename x and y in input data to avoid conflicts
    if 'x' in df_input_train_processed.columns:
        df_input_train_processed = df_input_train_processed.rename({'x': 'input_x', 'y': 'input_y'})


    # Merge processed input with training output
    # Use inner join to ensure we only keep frames present in both input and output
    train_data = df_input_train_processed.join(
        df_output_train,
        on=["game_id", "play_id", "nfl_id", "frame_id"],
        how="inner"
    )

    # Convert to pandas DataFrame for AutoGluon
    train_data_pd = train_data.to_pandas()

    # Ensure target columns are named 'x' and 'y' (these should be from the output df)
    # The merge should have kept the original names for df_output_train
    # Check for potential suffixes if join resulted in conflicts (though renaming input_x/input_y should prevent this)
    if 'x_right' in train_data_pd.columns: # If output x got renamed
         train_data_pd = train_data_pd.rename(columns={'x_right': 'x', 'y_right': 'y'})
         # Drop the original input x and y columns which will have '_left' suffix if not renamed before join
         if 'x_left' in train_data_pd.columns:
              train_data_pd = train_data_pd.drop(columns=['x_left', 'y_left'])
    elif 'x' not in train_data_pd.columns: # If output x was somehow dropped or not named 'x'
         print("Error: Target 'x' and 'y' not found after merge. Check merge logic.")
         # Depending on desired behavior, you might raise an error or use a fallback
         pass


    # Drop the original input x and y columns if they still exist and are not needed as features
    # Assuming input_x and input_y are kept as features
    # If you don't want input_x/input_y as features, drop them here:
    # train_data_pd = train_data_pd.drop(columns=['input_x', 'input_y'])


    # Verify unique column names before returning
    if len(set(train_data_pd.columns)) < len(train_data_pd.columns):
        duplicate_cols = [col for col in train_data_pd.columns if list(train_data_pd.columns).count(col) > 1]
        print(f"Error: Duplicate columns found after merge: {duplicate_cols}")
        # Attempt to handle common duplicates if not already handled
        # This might need more specific logic based on which columns are duplicated
        # For now, let's print and let AutoGluon potentially handle (though it raised error before)
        pass # Let AutoGluon handle the error after printing


    print(f"Prepared training data shape: {train_data_pd.shape}")
    print(f"Training data columns: {train_data_pd.columns.tolist()}")

    return train_data_pd

def prepare_test_data(test_input_path, test_template_path):
    """Prepare test data for inference"""
    print("Preparing test data for inference...")

    # Load test data
    df_test_in = pl.read_csv(test_input_path)
    test_template = pd.read_csv(test_template_path)

    print(f"Test input shape: {df_test_in.shape}")
    print(f"Test template shape: {test_template.shape}")

    # Apply feature engineering to test input data
    df_test_in_processed = process_data_for_autogluon(df_test_in)

    # Rename x and y in the processed test input data to match the feature names used in training
    # This is crucial for the model to find the 'input_x' and 'input_y' features
    if 'x' in df_test_in_processed.columns:
        df_test_in_processed = df_test_in_processed.rename({'x': 'input_x', 'y': 'input_y'})


    # Merge processed test input with test template to get the frames to predict
    # Use left join to keep all frames from the test template
    test_data = pl.from_pandas(test_template).join(
        df_test_in_processed,
        on=['game_id', 'play_id', 'nfl_id', 'frame_id'],
        how='left'
    )

    # Add 'player_to_predict' column to test data.
    # In the test data, the 'player_to_predict' is the player specified by the unique game_id, play_id, and nfl_id combination.
    # Since the test template only contains rows for the player to predict for each frame,
    # all rows in the merged test_data should correspond to the player to predict.
    test_data = test_data.with_columns(pl.lit(True).alias("player_to_predict"))


    # Convert to pandas DataFrame for AutoGluon
    test_data_pd = test_data.to_pandas()

    # Ensure all necessary columns are present for prediction (match training features)
    # This will be handled by AutoGluon automatically, but good to keep in mind.
    # Fill missing values created by the merge or feature engineering
    test_data_pd = test_data_pd.fillna(-1) # Simple fillna, consider more sophisticated methods


    print(f"Prepared test data shape: {test_data_pd.shape}")
    print(f"Test data columns: {test_data_pd.columns.tolist()}")

    return test_data_pd


def main():
    """Main training and inference function using AutoGluon"""
    print("NFL Big Data Bowl 2026 - AutoGluon Training and Inference Pipeline")
    print("=" * 60)

    # Paths
    train_input_path = os.path.join(DATA_DIR, "train/input_2023_w01.csv") # Use specific training data
    train_output_path = os.path.join(DATA_DIR, "train/output_2023_w01.csv") # Use specific training data
    test_input_path = os.path.join(DATA_DIR, "test_input.csv") # Use /content/ for test_input.csv
    test_template_path = os.path.join(DATA_DIR, "test.csv") # Use /content/ for test.csv
    # MODELS_DIR is set to /kaggle/working/autogluon-models/
    # OUTPUT_DIR is set to /kaggle/working/

    # --- Training ---
    print("\n--- Training Phase ---")

    # Load training data
    print("Loading training data...")
    df_input_train = pl.read_csv(train_input_path)
    df_output_train = pl.read_csv(train_output_path)
    print(f"Training input shape: {df_input_train.shape}")
    print(f"Training output shape: {df_output_train.shape}")


    # Prepare training data
    train_data_pd = prepare_training_data(df_input_train, df_output_train)

    # Define target columns
    target_x = 'x'
    target_y = 'y'

    # Train AutoGluon predictor for x
    print(f"\nTraining AutoGluon predictor for {target_x}...")
    predictor_x = TabularPredictor(label=target_x, path=os.path.join(MODELS_DIR, 'x_predictor'), eval_metric='root_mean_squared_error').fit(
        train_data_pd,
        time_limit=360, # Set a time limit for training (in seconds)
        presets='best_quality' # Use a preset for better quality models
    )
    print(f"Training summary for {target_x}:")
    print(predictor_x.fit_summary())


    # Train AutoGluon predictor for y
    print(f"\nTraining AutoGluon predictor for {target_y}...")
    predictor_y = TabularPredictor(label=target_y, path=os.path.join(MODELS_DIR, 'y_predictor'), eval_metric='root_mean_squared_error').fit(
        train_data_pd,
        time_limit=360, # Set a time limit for training (in seconds)
        presets='best_quality' # Use a preset for better quality models
    )
    print(f"Training summary for {target_y}:")
    print(predictor_y.fit_summary())

    print("\n--- Training Phase Complete ---")

    # --- Inference ---
    print("\n--- Inference Phase ---")

    # Load trained AutoGluon models (from the working directory where they were saved)
    print(f"Loading trained AutoGluon models from: {MODELS_DIR}")
    # AutoGluon's load function should handle finding the model within the specified path
    predictor_x = TabularPredictor.load(os.path.join(MODELS_DIR, 'x_predictor'))
    predictor_y = TabularPredictor.load(os.path.join(MODELS_DIR, 'y_predictor'))


    # Prepare test data
    test_data_pd = prepare_test_data(test_input_path, test_template_path)

    # Define features for prediction (excluding id columns) - should match training features
    # AutoGluon handles feature matching automatically during prediction,
    # but it's good practice to be aware of the feature set used.
    features_to_predict = [col for col in test_data_pd.columns if col not in ['game_id', 'play_id', 'nfl_id', 'frame_id', 'id']]
    # Ensure features used for prediction are a subset of features used in training if needed
    # For simplicity with AutoGluon, we'll let it handle feature alignment based on the loaded predictor


    # Make predictions
    print("Making predictions with AutoGluon...")

    # AutoGluon automatically handles feature selection based on the trained model
    predictions_x = predictor_x.predict(test_data_pd)
    predictions_y = predictor_y.predict(test_data_pd)


    # Create submission file
    print("Creating submission file...")
    submission_df = pd.DataFrame({
        'id': test_data_pd['id'],
        'x': predictions_x,
        'y': predictions_y
    })

    # Save submission
    submission_path = os.path.join(OUTPUT_DIR, "submission.csv")
    submission_df.to_csv(submission_path, index=False)
    print(f"Saved submission.csv ({len(submission_df)} rows)")

    # Print summary
    print(f"\n{'='*60}")
    print("AUTOGLUON TRAINING AND INFERENCE COMPLETE!")
    print(f"{'='*60}")
    print(f"Submission files saved to: {OUTPUT_DIR}")
    print(f"Final submission shape: {submission_df.shape}")
    print(f"Sample predictions:")
    print(submission_df.head())

if __name__ == "__main__":
    main()
```
